**Final Exam Study Plan Tool – Requirements & Prototype & Pitch**

## 一、需求定义（PRD 简版）【10–20 分钟】

**I. Requirement Definition (PRD – Simplified) \[10–20 mins]**

### 1. 目标用户

* 大学生及备考学生（尤其是面临多科目期末考试的本科生）
* 需求：高效规划复习时间，按优先级分配精力

**Target Users**

* College students and exam-prep learners (especially undergraduates with multiple final exams).
* Needs: Efficiently plan study time and allocate effort based on priority.

---

### 2. 用户核心任务

* 根据可用复习天数和待考科目
* 生成按知识点重要性排序的每日复习计划
* 每日计划包含：时间分配、知识点、复习指南、辅助资源

**User Core Tasks**

* Based on available study days and subjects to review.
* Generate a daily study plan sorted by importance of topics.
* Each day includes: time allocation, key topics, study guide, and learning resources.

---

### 3. 输入/输出

* **输入**：

  * 复习总天数（1–14天）
  * 待复习科目（多选，预设 4 科）
* **输出**：

  * 网页展示的结构化复习计划（按天划分）
  * 可下载 PDF（保留颜色和复习指南）
  * 可复制文本清单（便于笔记/文档）

**Input / Output**

* **Input**:

  * Total number of study days (1–14 days).
  * Subjects to review (multi-select, 4 preset subjects).
* **Output**:

  * Structured study plan displayed on webpage (organized by day).
  * Downloadable PDF (preserving colors and study guide).
  * Copyable text checklist (for notes/documents).

---

| **MVP 必须功能**  | **可扩展功能**       |
| ------------- | --------------- |
| 多科目选择         | 自定义科目添加         |

| 复习指南（含资源链接）   | 社交分享 / 计划卡片     |

## 三、补充说明 & 创意延伸【40–50 分钟】

**III. Supplement & Creative Extensions \[40–50 mins]**

### 1. 需要后端的部分

* 用户系统（保存历史计划、偏好） → 数据库存储
* 高级 PDF 生成（复杂排版/水印/批量生成） → Node.js + PDFKit
* 邮件推送 → 后端接口调用邮件服务
* 学习资源链接校验 → 后端定时任务（Python 脚本）

**Backend Needed For**

* User system (save history & preferences) → Database storage
* Advanced PDF generation (complex layout/watermark/batch) → Node.js + PDFKit
* Email delivery → Backend API + email service
* Resource link validation → Scheduled backend job (Python script)

---

### 2. 可由 AI Agents 自动化

* 计划优化：分析历史完成情况 → 自动调整知识点分配
* 资源推荐：根据科目/知识点 → 自动爬取最新学习资源
* 语言适配：自动翻译更多语言
* 错题关联：上传错题 → AI 自动匹配相关知识点

**AI Agent Automation**

* Plan optimization: analyze past completion → auto-adjust topics allocation
* Resource recommendation: auto-fetch updated learning resources
* Language adaptation: auto-translate into more languages
* Wrong-note linking: upload mistakes → AI matches related topics

## 四、3 分钟 Pitch 展示脚本【50–60 分钟】

**IV. 3-Minute Pitch Script \[50–60 mins]**

### 开场（30秒）

“大家好！期末复习时间紧，科目多，没头绪？输入复习天数和科目，我们的工具 1 秒生成按重要性排序的个性化复习计划，还可下载 PDF 或复制到笔记！”

**Opening (30s)**
“Hi everyone! Finals are stressful, too many subjects and too little time? Just enter your study days and subjects — our tool generates a personalized, prioritized study plan in 1 second, downloadable as PDF or copyable to notes!”

---

### 功能演示（1分钟）

* 选择 3 天复习高数和英语 → 点击生成
* 每日计划显示重点知识（红=高频考点，黑=基础内容，灰=拓展）
* 点击知识点展开复习指南和学习链接
* 下载 PDF 或复制文本到笔记

**Demo (1 min)**

* Select 3 days for Math & English → click generate
* Daily plan shows key topics (Red = high-frequency, Black = core, Grey = extra)
* Expand topics for study guide & resource links
* Download PDF or copy text to notes

---

### 技术思路（1分钟）

* 前端：HTML + JS + Tailwind
* 核心逻辑：动态筛选知识点（天数少只保留高优先级）、难科优先分配天数
* 可扩展：后端用户系统、AI 个性化推荐

**Tech Idea (1 min)**

* Frontend: HTML + JS + Tailwind
* Core logic: dynamically filter topics (shorter plans = only high-priority topics), harder subjects get more time.
* Extensible: backend user system, AI personalized recommendations.

---

### 收尾（30秒）

“无论 3 天突击还是 2 周系统复习，这个工具都能帮你高效规划。目前已上线，可直接体验！”

**Closing (30s)**
“Whether it’s 3-day cramming or 2-week systematic prep, this tool helps you plan efficiently. Already online and ready to try!”
